### Embedding del texto y almacenamiento en ChromaDB:

```python
import chromadb

# Creamos el objeto de la base de datos
bd_chroma = chromadb.Client()
```

```python
# Eliminar una coleccion
# chroma_client.delete_collection(name=f"Diccionario_De_Psicoanalisis")

# Eliminar todas las colecciones
# for nombre in textos_cortados.keys():
#   coleccion = chroma_client.delete_collection(name=f"{nombre}")
```

```python
import tensorflow_text
import tensorflow_hub as hub
from chromadb import EmbeddingFunction, Documents, Embeddings

class MyEmbeddingFunction(EmbeddingFunction):

    def __init__(self):
        super().__init__()

        # Cargar el modelo Universal Sentence Encoder multilingüe desde TensorFlow Hub
        self.embed = hub.load("https://www.kaggle.com/models/google/universal-sentence-encoder/frameworks/TensorFlow2/variations/multilingual/versions/2")

    def __call__(self, input: Documents) -> Embeddings:
        embeddings = []
        batch_size = 32  # Ajusta este valor según la capacidad de tu GPU
        for i in range(0, len(input), batch_size):
            batch = input[i:i + batch_size]
            embeddings_batch = self.embed(batch).numpy().tolist()
            embeddings.extend(embeddings_batch)
        return embeddings

# Crear una instancia de la función de embedding personalizada
funcion_embedding = MyEmbeddingFunction()
```

```python
# Aca vamos a guardar los embeddings que generemos
embeddings_textos_cortados = {}

# Por cada guia con split ya hecho
for nombre, texto in textos_cortados.items():
  # Creamos los embeddings del documento
  embeddings_textos_cortados[nombre] = funcion_embedding(textos_cortados[nombre])
  # Creamos una coleccion perteneciente a ese libro
  coleccion = bd_chroma.create_collection(name=f"{nombre}", metadata={"hnsw:space": "cosine"})
  # Agregamos:
  coleccion.add(
      embeddings=embeddings_textos_cortados[nombre],
      # Los splits
      documents=texto,
      # Los metadatos
      metadatas = [ {'Libro': nombre} for _ in range(len(texto)) ],
      # Como id que numero de chunk es dicho split
      ids=[f'parte {str(x)}' for x in range(len(texto))]
  )
```

```python
# Funcion para realizar la busqueda semantica, recibe el string correspondiente al nombre de la coleccion y la query
def busqueda_semantica(nombre, query):

  # Cantidad de resultados que traera
  cantidad_resultados=2
  # A partir del nombre del libro se obtiene la coleccion correspondiente
  coleccion = bd_chroma.get_collection(name=nombre, embedding_function=funcion_embedding)
  # Busqueda en si.
  resultado_query = coleccion.query(
      query_texts=query,
      n_results=cantidad_resultados,
      # where={"texto": {"$eq": 'Diccionario_De_Psicoanalisis'}},
      # where_document={"$contains":"Freud"},
      # Devuelve distancia (coseno seteada arriba), metadata (nombre del libro) y los documentos mas cercanos (chunks correspondientes)
      include=['distances', 'metadatas', 'documents']
  )
  # Formateo de los resultados para luego pasar al contexto del LLM.
  resultado_limpio=f'Libro: {nombre}\n\n'
  for i in range(cantidad_resultados):
    resultado_limpio += f'{resultado_query["ids"][0][i]}:\n'
    resultado_limpio += f'Similitud coseno: {resultado_query["distances"][0][i]}\n'
    resultado_limpio += f'{resultado_query["documents"][0][i]}:\n\n'

  # Devuelve el resultado formateado en un string
  return resultado_limpio
```

### Funcion de consulta tabular:

```python
from fuzzywuzzy import process

#Se utiliza una busqueda flexible por similitud para que sea tolerante en la busqueda

def busqueda_tabular(query):
    resultado_años = None
    resultado_autor = None
    resultado_libro = None

    if type(query) == int:
        resultado_años = Años.query(f"Año == {query}")
    else:
        autores = Años['Autor'].unique()
        libros = Años['Titulo'].unique()

        mejor_autor, score_autor = process.extractOne(query, autores)
        mejor_libro, score_libro = process.extractOne(query, libros)

        # Ajusta el umbral de similitud según sea necesario (0-100)
        umbral_similitud = 70

        if score_autor >= umbral_similitud:
            resultado_autor = Años.query(f"Autor == '{mejor_autor}'")

        if score_libro >= umbral_similitud:
            resultado_libro = Años.query(f"Titulo == '{mejor_libro}'")

    return resultado_años, resultado_autor, resultado_libro
```

### Creacion de las querys para traer la informacion de cada autor entre 5 posibles desde Wikidata:

```python
from SPARQLWrapper import SPARQLWrapper, JSON
import sys
import pandas as pd
from typing import List, Dict

class consulta_sparql:
    """
    A class that can be used to query data from Wikidata using SPARQL and return the results as a Pandas DataFrame or a list
    of values for a specific key.
    """
    def __init__(self, query: str):
        """
        Initializes the WikiDataQueryResults object with a SPARQL query string.
        :param query: A SPARQL query string.
        """
        self.user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
        self.endpoint_url = "https://query.wikidata.org/sparql"
        self.sparql = SPARQLWrapper(self.endpoint_url, agent=self.user_agent)
        self.sparql.setQuery(query)
        self.sparql.setReturnFormat(JSON)

    def __transform2dicts(self, results: List[Dict]) -> List[Dict]:
        """
        Helper function to transform SPARQL query results into a list of dictionaries.
        :param results: A list of query results returned by SPARQLWrapper.
        :return: A list of dictionaries, where each dictionary represents a result row and has keys corresponding to the
        variables in the SPARQL SELECT clause.
        """
        new_results = []
        for result in results:
            new_result = {}
            for key in result:
                new_result[key] = result[key]['value']
            new_results.append(new_result)
        return new_results

    def _load(self) -> List[Dict]:
        """
        Helper function that loads the data from Wikidata using the SPARQLWrapper library, and transforms the results into
        a list of dictionaries.
        :return: A list of dictionaries, where each dictionary represents a result row and has keys corresponding to the
        variables in the SPARQL SELECT clause.
        """
        results = self.sparql.queryAndConvert()['results']['bindings']
        results = self.__transform2dicts(results)
        return results

    def load_as_dataframe(self) -> pd.DataFrame:
        """
        Executes the SPARQL query and returns the results as a Pandas DataFrame.
        :return: A Pandas DataFrame representing the query results.
        """
        results = self._load()
        return pd.DataFrame.from_dict(results)
```

```python
# Funcion que define la plantilla jinja
def plantilla_instruccion_zephyr(messages, add_generation_prompt=True):
  template_str = "{% for message in messages %}"
  template_str += "{% if message['role'] == 'user' %}"
  template_str += "<|user|>{{ message['content'] }}</s>\n"
  template_str += "{% elif message['role'] == 'assistant' %}"
  template_str += "<|assistant|>{{ message['content'] }}</s>\n"
  template_str += "{% elif message['role'] == 'system' %}"
  template_str += "<|system|>{{ message['content'] }}</s>\n"
  template_str += "{% else %}"
  template_str += "<|unknown|>{{ message['content'] }}</s>\n"
  template_str += "{% endif %}"
  template_str += "{% endfor %}"
  template_str += "{% if add_generation_prompt %}"
  template_str += "<|assistant|>\n"
  template_str += "{% endif %}"
  # Crear un objeto de plantilla con la cadena de plantilla
  template = Template(template_str)
  # Renderizar la plantilla con los mensajes proporcionados
  return template.render(messages=messages, add_generation_prompt=add_generation_prompt)

# Aquí hacemos la llamada el modelo
def conexion_llm(prompt: str, max_new_tokens: int = 768) -> None:
  try:
    # Tu clave API de Hugging Face
    api_key = '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'

    # URL de la API de Hugging Face para la generación de texto
    api_url = "https://api-inference.huggingface.co/models/HuggingFaceH4/zephyr-7b-beta"

    # Cabeceras para la solicitud
    headers = {"Authorization": f"Bearer {api_key}"}

    # Datos para enviar en la solicitud POST
    # Sobre los parámetros: https://huggingface.co/docs/transformers/main_classes/text_generation
    data = {
      "inputs": prompt,
      "parameters": {
        "max_new_tokens": max_new_tokens,
        "temperature": 0.01,
      }
    }

    # Realizamos la solicitud POST
    response = requests.post(api_url, headers=headers, json=data)

    # Extraer respuesta
    respuesta = response.json()[0]["generated_text"][len(prompt):]

    return respuesta

  except Exception as e:
    print(f"Error durante la conexion con el LLM: Error {e}")

# Clasificador de categoria para seleccionar la base de datos basado en LLM:
def clasificador_base_datos(prompt_inicial: str):

  PLANTILLA_CLASIFICACION_BASE_DATOS = (
    "Dadas las siguientes categorias clasifica el texto en unicamente una categoria.\n"
    "Categorias:\n"
     "1\n"
     "2\n"
     "3\n"
     "4\n"
     '''La categoria 1 corresponde a teoría sobre psicoanalisis, preguntas o enunciados sobre conceptos, como pulsión, goce, etc.\n'''
     '''La categoria 2 corresponde a fechas relevantes sobre publicaciones de libros de psicoanalisis de Sigmund Freud y Jacqes Lacan, etc\n'''
     '''La categoria 3 corresponde a información biográfica de Sigmund Freud y Jacques Lacan, Jung, Klein o Winnicott, padres del psicoanalisis etc\n'''
     "La categoria 4 corresponde a ninguna de las anteriores\n"
     "Pregunta: {prompt_inicial}\n"
     "Respuesta: "
  )

  mensaje = [
    {
      "role": "system",
      "content": '''Eres un clasificador de texto en categorias.''',
    },
    {"role": "user", "content": PLANTILLA_CLASIFICACION_BASE_DATOS.format(prompt_inicial=prompt_inicial)},
  ]

  prompt_plantilla = plantilla_instruccion_zephyr(mensaje)

  clasificacion_base_datos = conexion_llm(prompt_plantilla)

  return clasificacion_base_datos


# Funcion que utiliza el LLM para limpiar el prompt de consulta a la base de datos semantica:
def limpiar_prompt_guia(prompt_inicial: str):

  PLANTILLA_LIMPIAR_PROMPT_GUIA = (
     "Cadena de texto: {prompt_inicial}.\n"
  )

  mensaje = [
    {
      "role": "system",
      "content": '''Eres un asistente que recibe una cadena de texto del usuario que tratara acerca de informacion referente al
      psicoanalisis
      .\n'''
     '''Tu salida debe ser unicamente el texto formateado\n''',
    },
    {"role": "user", "content": PLANTILLA_LIMPIAR_PROMPT_GUIA.format(prompt_inicial=prompt_inicial)},
  ]

  prompt_plantilla = plantilla_instruccion_zephyr(mensaje)

  prompt_limpio_llm = conexion_llm(prompt_plantilla)

  # # Ultimos retoques que dejan el prompt limpio
  # prompt_limpio = prompt_limpio_llm.split('Cadena de texto sin el nombre del libro: ')[1].split('\n')[0]

  return prompt_limpio_llm


# Funcion para generar la respuesta final (compartida para todas las categorias)
def generar_respuesta_guia(query_str: str, contexto: str):

  TEXT_QA_PROMPT_TMPL = (
    "\n\n------------------------------------------------------------------------\n"
    "Informacion de contexto:\n"
    "{context_str}\n"
    "------------------------------------------------------------------------\n"
    "Pregunta: {query_str}\n"
    "Respuesta: \n\n"
  )


  messages = [
    {

     "role": "system",
      "content": '''Responde en ESPAÑOL.\n
Eres un asistente especializado en psicoanalisis que responde amablemente preguntas guiandose unicamente por la
informacion de contexto pero respondes como si esta no existiera y fueras tu quien lo sabia de antemano.\n
'''
      "Responde lo mas corto posible.\n"
      "Responde sin utilizar conocimiento previo.\n"

    },

    {"role": "user",
     "content": TEXT_QA_PROMPT_TMPL.format(context_str=contexto, query_str=query_str)},
  ]

  prompt_final = plantilla_instruccion_zephyr(messages)

  # print('----------------------------------------------------------------------------------------')
  # print('')
  # print('Control Prompt que recibe el LLM:')
  # print('')
  # print(prompt_final)
  # print('----------------------------------------------------------------------------------------')
  # print('')

  respuesta_final = conexion_llm(prompt_final)

  return respuesta_final

# Funcion que deriva la consulta a la base de datos correspondiente para buscar el contexto y luego genera la respuesta
def respuesta_chatbot(libro, pregunta: str, categoria: str, referencia=None):

  if categoria == '1': resultado_bdd = busqueda_semantica(libro, pregunta)
  elif categoria == '2': resultado_bdd = busqueda_tabular(libro)
  elif categoria == '3': resultado_bdd = consulta_sparql(libro)._load()[:2]

  respuesta = generar_respuesta_guia(pregunta, resultado_bdd)

  return respuesta
```

Prueba de clasificador basado en LLM en 2 pasos, primero con el modelo Zephir y luego Zero-Shot para extraer el número de la categoría en el formato deseado.

```python
# Clasificador zero-shot de HugginFace utilizado tanto como segunda etapa del clasificador de categorias como
# clasificador de libros y autores
clasificador_zero_shot = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")

# Etiquetas para la segunda etapa del clasificador de categorias
etiquetas_base_datos = ['1', '2', '3', '4']

# Dos grupos de etiquetas para el clasificador de libros
etiquetas_libros = ["Estructuras_clinicas_y_psicoanalisis", "Diccionario_De_Psicoanalisis","lacan_seminario_20"]

etiquetas_autor = ['Sigmund_Freud', 'Jacques_Lacan', 'Melanie_Klein', 'Donald_Winnicott', 'Gustav_Jung']
```

```python
# Etiquetas de las bases de datos
categorias = [1, 2, 3, 4]

# Contadores para el número de aciertos
aciertos_zero_shot = 0

# Diccionarios para contar aciertos y total por categoría
aciertos_por_categoria = {etiqueta: 0 for etiqueta in categorias}
total_por_categoria = {etiqueta: 0 for etiqueta in categorias}

# Total de preguntas en el dataset
total_preguntas = len(dataset)

# Recorrer todos los elementos del dataset
for idx, (categoria_real, pregunta) in enumerate(dataset):
    clasificacion_bdd_llm = clasificador_base_datos(pregunta)
    clasificacion_bdd_zero = clasificador_zero_shot(clasificacion_bdd_llm, categorias, multi_label=False)

    # Obtener la etiqueta predicha por el clasificador zero-shot
    categoria_predicha_zero_shot = clasificacion_bdd_zero['labels'][0]

    # Contabilizar la pregunta en la categoría correspondiente
    total_por_categoria[categoria_real] += 1

    if int(categoria_predicha_zero_shot) == categoria_real:
        aciertos_zero_shot += 1
        aciertos_por_categoria[categoria_real] += 1

    # Imprimir resultados para cada pregunta
    print(f"Pregunta {idx + 1}: {pregunta}")
    print(f"Categoría real: {categoria_real}")
    print(f"Clasificación Zero-Shot: {categoria_predicha_zero_shot}")

# Calcular porcentaje de aciertos
porcentaje_aciertos_zero_shot = (aciertos_zero_shot / total_preguntas) * 100

print(f"Porcentaje de aciertos del clasificador Zero-Shot: {porcentaje_aciertos_zero_shot:.2f}%")

# Calcular y mostrar el porcentaje de aciertos por categoría
print("\nPorcentaje de aciertos por categoría:")
for categoria, total in total_por_categoria.items():
    if total > 0:
        porcentaje_aciertos_categoria = (aciertos_por_categoria[categoria] / total) * 100
        print(f"Categoría {categoria}: {porcentaje_aciertos_categoria:.2f}% (Aciertos: {aciertos_por_categoria[categoria]} / Total: {total})")
    else:
        print(f"Categoría {categoria}: No hay preguntas en esta categoría.")
```

El porcentaje general de aciertos es superior al clasificador de embbedings. Como contraparte el tiempo de ejecución es mayor, pero despreciable al no requerirse un uso intensivo.
